```python
choices=["coffee", "tea", "pastries" , "Coffee", "Tea", "Pastries"]
Coffee_Choices = {"Espresso": (1.50) ,
                  "Americano" : (2.00) , 
                  "Caffe latte" : (2.25) , 
                  "Mocha" : (2.25) ,
                  "Cappucino" : (2.50) ,
                  "Macchiato" : (2.50)}

Tea_Choices = {"White tea" : 1.50 ,
               "Green tea" : 2.00 , 
               "Oolong tea" :2.25 ,
               "Black tea" : 2.25 ,
               "Earl Grey" : 2.5 ,
               "Matcha": 2.50}

Pastries_Choices = {"Croissant" : 5.50 , 
                    "Cinnamon roll" : 4.50 , 
                    "Banana bread" : 4.25 , 
                    "Pumpkin bread" : 3.75, 
                    "Sticky bun" : 4.25,
                    "Apple turnover" : 3.75}
```

```python
def direct_order():
    cost = 0
    cont= "y"
    while cont=="y":
        my_order = []
        total= []
        item = input("What do you want today?")
        quantity = int(input("Please specify number of items"))
        if item in Coffee_Choices:
            print (f" item = {item}, \t [item] = {Coffee_Choices[item]}") 
            price = Coffee_Choices[(f"{item}")]
            my_order.append({item})
            print("Your order is",my_order)
            cost_coffee = (quantity * price)
            cost+=cost_coffee
            total.append({cost})
            tax = cost*0.06
            print ("You pay $",  (cost+tax) , "(Tax included)")
        if item in Tea_Choices:
            print (f" item = {item}, \t [item] = {Tea_Choices[item]}") 
            price = Tea_Choices[(f"{item}")]
            my_order.append({item})
            print("Your order is",my_order)
            cost_tea = (quantity * price)
            cost+=cost_tea
            total.append({cost})
            tax = cost*0.06
            print ("You pay $",  (cost+tax) , "(Tax included)")
        if item in Pastries_Choices:
            print (f" item = {item}, \t [item] = {Pastries_Choices[item]}") 
            price = Pastries_Choices[(f"{item}")]
            my_order.append({item})
            print("Your order is",my_order)
            cost_pastries = (quantity * price)
            cost+=cost_pastries
            total.append({cost})
            tax = cost*0.06
            print ("You pay $",  (cost+tax) , "(Tax included)")
                       
        cont=input("Do you wish to continue ordering? (Y/N)")

 
   
```
